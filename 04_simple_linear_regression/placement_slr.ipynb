# Linear Regression from Scratch 
**Dataset:** Placement Data (`placement.csv`)  
**Goal:** Prove a fundamental understanding of how Simple Linear Regression works under the hood. We will build a custom Linear Regression class using pure mathematical formulas (calculating slope $m$ and intercept $b$) and compare its performance directly against `scikit-learn`'s industry-standard implementation.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score 
```

## 1. Building the Custom Algorithm
To build Simple Linear Regression from scratch, we need to calculate the line of best fit: $y = mx + b$.

The mathematical formulas used in this class are:
*   **Slope ($m$):**  $m = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}$
*   **Intercept ($b$):** $b = \bar{y} - m\bar{x}$

*Where $\bar{x}$ and $\bar{y}$ are the means of the training data.*

```python
class Linear_Regression:
    def __init__(self):
        self.m=None
        self.b=None

    def fit(self, X_train, y_train):
        num=0
        den=0
        X_mean=X_train.mean()
        y_mean=y_train.mean()
        for i in range(X_train.shape[0]):
            num+=(X_train[i]-X_mean)*(y_train[i]-y_mean)
            den+=(X_train[i]-X_mean)*(X_train[i]-X_mean)
        self.m=num/den
        self.b=y_mean-(self.m*X_mean)
        print(self.m)
        print(self.b)


    def predict(self, X_test):
        return self.m*X_test+self.b
```

```python
df=pd.read_csv('placement.csv')
```

## 2. Training the Custom Model
We will extract our features ($X$) and target ($y$), split them into training and testing sets, and then fit our custom `Linear_Regression` class. We extract the values as 1D numpy arrays because our custom loop expects flat lists of numbers.

```python
X=df.iloc[:, 0].values
y=df.iloc[:, 1].values
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
lr=Linear_Regression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
mae_custom = mean_absolute_error(y_test, y_pred)
mse_custom = mean_squared_error(y_test, y_pred)
rmse_custom = np.sqrt(mse_custom)
r2_custom = r2_score(y_test, y_pred)
```

## 3. Scikit-Learn Implementation
Now, we implement the exact same model using `sklearn.linear_model.LinearRegression`. 
*Note: Scikit-Learn expects 2D arrays for its features (even for simple linear regression), so we must reshape $X$ before fitting the model.*

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_sk=LinearRegression()
```

```python
X_train_2d = X_train.reshape(-1, 1)
X_test_2d = X_test.reshape(-1, 1)
```

```python
lr_sk.fit(X_train_2d, y_train)
```

```python
print("\nSklearn Model:")
print("Slope (m):", lr_sk.coef_[0])
print("Intercept (b):", lr_sk.intercept_)
```

```python
y_pred_sk=lr_sk.predict(X_test_2d)
```

## 4. Model Verification & Comparison
If our custom math is correct, the slope ($m$), intercept ($b$), and all evaluation metrics (MAE, MSE, RMSE, $R^2$) should perfectly match Scikit-Learn's highly optimized output. 
Let's compare the results side-by-side!

```python
mae_sk = mean_absolute_error(y_test, y_pred_sk)
mse_sk = mean_squared_error(y_test, y_pred_sk)
rmse_sk = np.sqrt(mse_sk)
r2_sk = r2_score(y_test, y_pred_sk)
```

```python
print("\nModel Comparison:")
print(f"MAE  -> Custom: {mae_custom:.4f} | Sklearn: {mae_sk:.4f}")
print(f"MSE  -> Custom: {mse_custom:.4f} | Sklearn: {mse_sk:.4f}")
print(f"RMSE -> Custom: {rmse_custom:.4f} | Sklearn: {rmse_sk:.4f}")
print(f"R2   -> Custom: {r2_custom:.4f} | Sklearn: {r2_sk:.4f}")
```
