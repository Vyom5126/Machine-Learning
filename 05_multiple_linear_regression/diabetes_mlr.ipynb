# Multiple Linear Regression from Scratch 
**Dataset:** Diabetes Dataset (`sklearn.datasets.load_diabetes`)  
**Goal:** Prove a fundamental understanding of how Multiple Linear Regression works under the hood. We will build a custom Multiple Linear Regression class using pure mathematical formulas (calculating coefficients $\mathbf{w}$ and intercept $b$ via the Normal Equation) and compare its performance directly against `scikit-learn`'s industry-standard implementation.

```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

## 1. Building the Custom Algorithm
To build Multiple Linear Regression from scratch, we need to find the best-fit hyperplane: $y = \mathbf{X}\mathbf{w} + b$.

We solve this analytically using the **Normal Equation**, which gives the exact closed-form solution:
*   **Coefficients ($\mathbf{w}$):** $\mathbf{w} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\mathbf{y}$

To handle the intercept cleanly, we prepend a column of ones to $\mathbf{X}$ so the bias term $b$ is absorbed into $\mathbf{w}$ as $w_0$.

*Where $\mathbf{X}$ is the feature matrix and $\mathbf{y}$ is the target vector of training data.*

```python
class Linear_Regression:
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None

    def fit(self, X_train, y_train):
        X_train=np.insert(X_train, 0, 1, axis=1)

        betas=np.linalg.inv(np.dot(X_train.T, X_train)).dot(X_train.T).dot(y_train)
        self.intercept_=betas[0]
        self.coef_=betas[1:]

    def predict(self, X_test):
        y_pred=np.dot(X_test, self.coef_) + self.intercept_
        return y_pred
        
```

## 2. Scikit-Learn Implementation
We implement the model using `sklearn.linear_model.LinearRegression`. 
*Note: Scikit-Learn's `LinearRegression` natively handles 2D feature matrices, so no reshape is needed for MLR — the data is already in the correct shape.*

```python
X,y=load_diabetes(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred_sk=reg.predict(X_test)
```

```python
mae_sk  = mean_absolute_error(y_test, y_pred_sk)
mse_sk  = mean_squared_error(y_test, y_pred_sk)
rmse_sk = np.sqrt(mse_sk)
r2_sk   = r2_score(y_test, y_pred_sk)
```

## 3. Training the Custom Model 
We will now fit our custom `Linear_Regression` class.

```python
lr=Linear_Regression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
mae_custom  = mean_absolute_error(y_test, y_pred)
mse_custom  = mean_squared_error(y_test, y_pred)
rmse_custom = np.sqrt(mse_custom)
r2_custom   = r2_score(y_test, y_pred)
```

## 4. Model Verification & Comparison
If our custom Normal Equation math is correct, the intercept ($b$), all coefficients ($\mathbf{w}$), and all evaluation metrics (MAE, MSE, RMSE, $R^2$) should perfectly match Scikit-Learn's highly optimized output. 
Let's compare the results side-by-side!

```python
print("\nModel Comparison:")
print(f"MAE  -> Custom: {mae_custom:.4f} | Sklearn: {mae_sk:.4f}")
print(f"MSE  -> Custom: {mse_custom:.4f} | Sklearn: {mse_sk:.4f}")
print(f"RMSE -> Custom: {rmse_custom:.4f} | Sklearn: {rmse_sk:.4f}")
print(f"R2   -> Custom: {r2_custom:.4f} | Sklearn: {r2_sk:.4f}")
```
